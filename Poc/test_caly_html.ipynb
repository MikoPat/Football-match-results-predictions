```python
dane
```

```python
def team_data(lista, nazwa):

    pattern = r'data-stat="([^"]+)">([^<]+)'

    # Use re.findall to extract all matching values
    matches = re.findall(pattern, lista)

    # Print the extracted values
    for match in matches:
        data_type, value = match
        dane[nazwa +data_type] = value
```

```python
team_data(home_team_summary, 'home_team_summary_')
team_data(home_team_passes, 'home_team_passes_')
team_data(home_team_pass_type, 'home_team_pass_type_')
team_data(home_team_defensive, 'home_team_defensive_')
team_data(home_team_possesion, 'home_team_possesion_')
team_data(home_team_miscellaneous, 'home_team_miscellaneous_')
```

```python
team_data(away_team_summary, 'away_team_summary_')
team_data(away_team_passes, 'away_team_passes_')
team_data(away_team_pass_type, 'away_team_pass_type_')
team_data(away_team_defensive, 'away_team_defensive_')
team_data(away_team_possesion, 'away_team_possesion_')
team_data(away_team_miscellaneous, 'away_team_miscellaneous_')
```
